# Tax regime Comparison 


## Range : 0 - 50,00,000
- Amount of Tax 

```python
incomes = [i*(5000) for i in range(60,1001)] # income between 2,50,000 and 50,00,000

for age in [25,65,85]:

    old_tax = []
    new_tax = []
    newer_tax = []

    for i in incomes:
        old_tax.append((calculate_total_tax(i, "old", age)))
        new_tax.append((calculate_total_tax(i, "new", age)))
        newer_tax.append((calculate_total_tax(i, "newer", age)))

    # Create traces for each line
    trace1 = go.Scatter(
        x = [i/100000 for i in incomes],
        y = old_tax,
        mode='lines',
        name='old tax regime'
    )
    trace2 = go.Scatter(
        x = [i/100000 for i in incomes],
        y = new_tax,
        mode='lines',
        name='2020 budget tax regime'
    )
    trace3 = go.Scatter(
        x = [i/100000 for i in incomes],
        y = newer_tax,
        mode='lines',
        name='2023 budget tax regime'
    )

    # Combine the traces into a data list
    data = [trace1, trace2, trace3]

    # Create layout for the chart
    layout = go.Layout(
        title = f'Tax Comparison for Age - {age}',
        xaxis = dict(title='Incomes (in lakhs.)'),
        yaxis = dict(title='Tax Amount')
    )

    # Create the figure and plot the chart
    fig = go.Figure(data=data, layout=layout)
    fig.show()
```

- Tax-rate

```python
incomes = [i*(5000) for i in range(60,1001)] # income between 2,50,000 and 50,00,000

for age in [25,65,85]:
    
    old_tax = []
    new_tax = []
    newer_tax = []

    for i in incomes:
        old_tax.append((calculate_total_tax(i, "old", age)/i)*100)
        new_tax.append((calculate_total_tax(i, "new", age)/i)*100)
        newer_tax.append((calculate_total_tax(i, "newer", age)/i)*100)

    # Create traces for each line
    trace1 = go.Scatter(
        x = [i/100000 for i in incomes],
        y = old_tax,
        mode='lines',
        name='old tax regime'
    )
    trace2 = go.Scatter(
        x = [i/100000 for i in incomes],
        y = new_tax,
        mode='lines',
        name='2020 budget tax regime'
    )
    trace3 = go.Scatter(
        x = [i/100000 for i in incomes],
        y = newer_tax,
        mode='lines',
        name='2023 budget tax regime'
    )

    # Combine the traces into a data list
    data = [trace1, trace2, trace3]

    # Create layout for the chart
    layout = go.Layout(
        title = f'Tax Comparison for Age - {age}',
        xaxis = dict(title='Incomes (in lakhs.)'),
        yaxis = dict(title='Tax Rate (%)')
    )

    # Create the figure and plot the chart
    fig = go.Figure(data=data, layout=layout)
    fig.show()
```

```python
tax_rate_after_adding_exemptions = [calculate_total_tax(j-i, "old", 25)/j for i,j in zip(Exemptions,incomes)] # tax-rate on income after adding exemptions using old regime
tax_rate_new_regime = [calculate_total_tax(i, "newer", 25)/i for i in incomes]
tax_rate_old_regime = [calculate_total_tax(i, "old", 25)/i for i in incomes]
```
